Al optar por este enfoque, estamos aplicando un método imparcial y objetivo para seleccionar individuos de la población de estudio de la empresa de calzado.

Dada la población relativamente pequeña de 20 colaboradores en Creaciones Joselyne, el **muestreo aleatorio simple** permite una recolección eficiente de datos sin incurrir en complejidades innecesarias.

##### Redondear al Entero Más Cercano:

##### Para este analisis vamos a utilizar el Lenguaje de programacion **Python** y librerias estadisticas como _numpy, scipy y matplotlib_

```python
# Librerias necesarias para el analisis 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

```

```python
# Parámetros del estudio
nivel_confianza = 0.95
margen_error = 0.05
proporcion_poblacion = 0.5  # Puedes asumir 0.5 si no tienes información
tamanio_poblacion = 20
```

```python
# Cálculo del valor crítico Z para el nivel de confianza
z_valor = norm.ppf((1 + nivel_confianza) / 2)
```

```python
# Cálculo del tamaño de la muestra
tamanio_muestra = int((z_valor**2 * proporcion_poblacion * (1 - proporcion_poblacion)) / (margen_error**2))
```

```python
# Visualización de resultados
print(f'Tamaño de la muestra necesario: {tamanio_muestra}')
```

```python
# Generación de una gráfica que muestra la distribución normal estándar
x = np.linspace(-3, 3, 100)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Distribución Normal Estándar')
plt.fill_between(x, y, where=[(xi >= -z_valor) and (xi <= z_valor) for xi in x], color='red', alpha=0.3, label='Área de Confianza')

plt.title(f'Distribución Normal con Nivel de Confianza del {nivel_confianza}')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parámetros del estudio
nivel_confianza = 0.95
margen_error = 0.05
proporcion_poblacion = 0.5  # Puedes asumir 0.5 si no tienes información
tamanio_poblacion = 20

# Cálculo del valor crítico t para el nivel de confianza (usando N-1 grados de libertad)
grados_libertad = tamanio_poblacion - 1
t_valor = t.ppf((1 + nivel_confianza) / 2, grados_libertad)

# Cálculo del tamaño de la muestra con desviación estándar muestral
tamanio_muestra = int((t_valor**2 * proporcion_poblacion * (1 - proporcion_poblacion)) / (margen_error**2) * (tamanio_poblacion / (tamanio_poblacion - 1)))

# Redondear al entero más cercano
tamanio_muestra = max(tamanio_muestra, 1)

# Visualización de resultados
print(f'Tamaño de la muestra necesario: {tamanio_muestra}')

# Gráfico de la distribución t con área sombreada para el intervalo de confianza
x = np.linspace(-3, 3, 100)
y = t.pdf(x, df=grados_libertad)

plt.plot(x, y, label=f'Distribución t con {grados_libertad} grados de libertad')
plt.fill_between(x, y, where=[(xi >= -t_valor) and (xi <= t_valor) for xi in x], color='red', alpha=0.3, label='Área de Confianza')

plt.title(f'Distribución t con Nivel de Confianza del {nivel_confianza}')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()

```
